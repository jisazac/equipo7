Este notebook toma la base de datos geographic.csv y la lleva a la siguiente estructura:


| Barrio | Lon | Lat |
| --- | --- | --- |
| string | Float | Float |

```python
#Se importa Pandas
import pandas as pd

# Lectura del archivo local
df=pd.read_csv("C:\\Users\\jisaza\\Desktop\\DS4A_workspace\\Dataset\\geographic.csv",encoding="utf8")

#Se crea una columna label que va a contener numeros 
df["Label"]=df.index
#Recorro el vector 
for i in range(len(df)):
    #Si es par lo asigno a Lon (longitud)
    if df.iloc[i,-1]%2==0:
        df.iloc[i,-1]="Lon"
    #Si es impar lo asigno a Lat (latitud)    
    else:
        df.iloc[i,-1]="Lat"
```

```python
#Separo el dataframe en uno que contenga latitudes y en otro que contenga longitudes
df_lat=df[df["Label"]=="Lat"]
df_lon=df[df["Label"]=="Lon"]
```

```python
#Aplico un unstack a cada daframe y los renombramos con el sufijo "2"
df_lon2=df_lon.T.unstack(level=0)
df_lat2=df_lat.T.unstack(level=0)

#Al objeto unstack() lo convertimos en dataframe
dfc=pd.DataFrame(df_lat2)
dfc.columns=["Lat"]
dff=pd.DataFrame(df_lon2)
dff.columns=["Lon"]

#Tomamos las tuplas del multiindex y generamos la variable Barrio
dff["Barrio"]=dff.index
dff["Barrio3"]=dff["Barrio"].astype(str)
dff['A.x'], dff['Barrio'] = zip(*dff.Barrio)
dff2=dff.reset_index(level=[0,1])

dfc["Barrio"]=dfc.index
dfc["Barrio3"]=dfc["Barrio"].astype(str)
dfc['A.x'], dfc['Barrio'] = zip(*dfc.Barrio)
dfc2=dfc.reset_index(level=[0,1])


#Rejuntamos el dataframe de longitudes con el de latitudes para obtenr el resultado final
df_final=dff2.join(dfc2, lsuffix='_caller', rsuffix='_other')
df_final=df_final[["Barrio_caller","Lon","Lat"]]
df_final.columns=["Barrio","Lon","Lat"]

```
