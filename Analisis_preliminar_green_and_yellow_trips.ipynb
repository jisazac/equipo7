```python
###Green Trips
green_trips = pd.read_csv("green_trips.csv")
green_trips.head()
green_trips['pickup_datetime']
```

```python
green_trips['pickup_datetime_dt'] = green_trips['pickup_datetime'].apply(lambda x: datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S"))
```

```python
green_trips['dropoff_datetime_dt'] = green_trips['dropoff_datetime'].apply(lambda x: datetime.datetime.strptime(str(x), "%Y-%m-%d %H:%M:%S"))
```

```python
green_trips_pickup_YYYYMM = green_trips['pickup_datetime_dt'].apply(lambda x: str(x.year)+'-'+str(x.month))
```

```python
print(green_trips_pickup_YYYYMM.unique())
```

```python
green_trips_dropoff_YYYYMM = green_trips['dropoff_datetime_dt'].apply(lambda x: str(x.year)+'-'+str(x.month))
green_trips_dropoff_YYYYMM.unique()
```
